```python
plt.plot(proc_num, mpi_time[0] / mpi_time, c='g')
plt.xlabel("К-во процессов")
plt.ylabel("Ускорение")
plt.grid()
```

```python
import matplotlib.pyplot as plt
import numpy as np

proc_num = np.array([2, 3, 5, 6, 7, 9, 11, 12])

mpi_time_2 = np.array([0.375746, 0.192835, 0.123115, 0.101387, 0.0925251, 0.0744853, 0.0610149, 0.0607478]) # A(2 * 256 x 2 * 256)
mpi_time_8 = np.array([23.9473, 12.0645, 6.15773, 5.67671, 4.58541, 4.97305, 4.26259, 4.17454]) # A(8 * 256 x 8 * 256)
mpi_time_16 = np.array([192.934, 97.5703, 51.3151, 43.4204, 51.0154, 40.1665, 34.0327, 32.529]) # A(16 * 256 x 16 * 256)

# Calculate theoretical acceleration
theoretical_acceleration_2 = (proc_num - 1) / (1 + 10 * (2 + proc_num) / (2 * 256))
theoretical_acceleration_8 = (proc_num - 1) / (1 + 10 * (2 + proc_num) / (8 * 256))
theoretical_acceleration_16 = (proc_num - 1) / (1 + 10 * (2 + proc_num) / (16 * 256))
```

```python
plt.figure(figsize=[10, 10])
plt.title("Матричное умножение")
plt.plot(proc_num - 1, proc_num - 1, c='gray', label="y=x", linestyle ='--')
plt.plot(proc_num - 1, theoretical_acceleration_2, c='y', linestyle ='--', label="теоритические матрицы: (512x512)")
plt.plot(proc_num - 1, theoretical_acceleration_8, c='g', linestyle ='--', label="теоритические матрицы: (2048x2048)")
plt.plot(proc_num - 1, theoretical_acceleration_16, c='b', linestyle ='--', label="теоритические матрицы: (4096x4096)")
plt.plot(proc_num - 1, mpi_time_2[0] / mpi_time_2, c='y', label="матрицы: (512x512)")
plt.plot(proc_num - 1, mpi_time_8[0] / mpi_time_8, c='g', label="матрицы: (2048x2048)")
plt.plot(proc_num - 1, mpi_time_16[0] / mpi_time_16, c='b', label="матрицы: (4096x4096)")
plt.xlabel("К-во процессов (мастер не учитыв)")
plt.ylabel(r"Ускорение ($t_0 / t_i$)")
plt.legend()
plt.grid()
```

```python
plt.figure(figsize=[10, 10])
plt.title("Матричное умножение")
plt.plot(proc_num - 1, mpi_time_2, c='y', label="матрицы: (512x512)")
plt.plot(proc_num - 1, mpi_time_8, c='g', label="матрицы: (2048x2048)")
plt.plot(proc_num - 1, mpi_time_16, c='b', label="матрицы: (4096x4096)")
plt.xlabel("К-во процессов (мастер не учитыв)")
plt.ylabel("Время, с")
plt.legend()
plt.grid()
```
